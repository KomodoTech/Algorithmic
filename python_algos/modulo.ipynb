```python
import numpy as np
import matplotlib.pyplot as plt

fig_width = 10
fig_height = 5

q = 7 # b = aq + r

x_max_factor = 5

b = np.arange(-x_max_factor*q, x_max_factor*q + 1)
#print(x)
r = b % q

fig = plt.figure(figsize = (fig_width, fig_height))
plt.scatter(b, r)

plt.title(f"r Python %{q}")
plt.xlabel('aq + r')
plt.ylabel("(aq + r) % q = r")

plt.xlim([-x_max_factor*q, x_max_factor*q])
plt.ylim([-q, q])

plt.xticks(np.arange(-x_max_factor*q, x_max_factor*q + 1, step=q))
plt.show()
```

```python
a = b // q

fig = plt.figure(figsize = (fig_width, fig_height))
plt.scatter(b, a)

plt.title(f"a Python %{q}")
plt.xlabel('aq + r')
plt.ylabel("(aq + r) // q = a")

plt.xlim([-x_max_factor*q, x_max_factor*q])
plt.ylim([-q, q])

plt.xticks(np.arange(-x_max_factor*q, x_max_factor*q + 1, step=q))
plt.show()
```
